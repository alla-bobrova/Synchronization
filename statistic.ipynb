```python
import matplotlib.pyplot as plt
```

```python
#считали из файла
vse=[]
with open("S2.txt") as f:
    for line in f:
        vse.append([float(x) for x in line.split()])

vse1=[]
for x in range(len(vse)):
  vse1.extend(vse[x])
```

```python
print(min(vse1))
print(max(vse1))
```

```python
groups=['A1', 'A2', 'A3', 'A4',]

k1=0
k2=0
k3=0
k4=0

for i in vse1:
    if i<0.3: k4+=1
    if (i>0.3) and (i<=0.45): k3+=1
    if (i>0.45) and (i<=0.49): k2+=1
    if (i>0.49) and (i<=0.5): k1+=1

counts=[k1,k2,k3,k4]
print(counts)

```

```python
plt.bar(groups, counts, zorder=3)

# labels
for i in range(len(groups)):
    plt.text(i, counts[i], str(counts[i]), ha='center', va='bottom')

plt.grid(color='gray', alpha=0.5, zorder=0)

plt.show()
```
